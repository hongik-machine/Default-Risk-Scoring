```python
# ===============================
# FULL PREPROCESSING (with rich prints)
# Input : uci_credit_clean.csv (클리닝 결과)
# Steps : Split → OneHot/Scale(CT) → (옵션) SMOTE/UnderSample on train
# Save  : X_train_processed.csv, X_test_processed.csv, y_train.csv, y_test.csv
# Note  : 모든 fit/통계는 train에만. 리샘플링도 train에만. (누수 방지)
# ===============================

import os, numpy as np, pandas as pd, matplotlib.pyplot as plt, warnings
from sklearn.model_selection import train_test_split
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from imblearn.over_sampling import SMOTE
from imblearn.under_sampling import ClusterCentroids

warnings.filterwarnings("ignore")
plt.rcParams["figure.figsize"] = (5,4)

# ---- 토글 ----
APPLY_SMOTE = True               # SMOTE 오버샘플링 (train only)
APPLY_CLUSTER_CENTROIDS = False  # 언더샘플링 (train only)
TEST_SIZE = 0.25
RANDOM_STATE = 42

# ---- 1) 데이터 로드 ----
csv_path = "uci_credit_clean.csv"
if not os.path.exists(csv_path):
    try:
        from google.colab import files
        print("'uci_credit_clean.csv' 업로드하세요.")
        up = files.upload()
        if not up: raise FileNotFoundError("업로드 실패")
        csv_path = list(up.keys())[0]
    except Exception as e:
        raise FileNotFoundError(f"파일 없음: {e}")

df = pd.read_csv(csv_path)

# === 로드 결과 요약 출력 ===
print("DF shape:", df.shape)
print("dtype 분포:", df.dtypes.value_counts().to_dict())
print("결측치 총개수:", int(df.isna().sum().sum()))
print("\n[HEAD 3]")
display(df.head(3))

# 타깃 분포
if "target" not in df.columns:
    raise KeyError("'target' 컬럼이 없습니다.")
print("\ntarget 분포(건수):\n", df["target"].value_counts().to_string())
print("\ntarget 분포(비율):\n", (df["target"].value_counts(normalize=True)*100).round(2).astype(str) + "%")

# 명목형 기초 분포(상위 3카테고리만 표시)
for col in ["SEX","EDUCATION","MARRIAGE"]:
    if col in df.columns:
        vc = df[col].value_counts().head(10)
        print(f"\n[{col}] top10 분포:")
        print(vc.to_string())

# ---- 2) 컬럼 그룹 ----
categorical = ["SEX", "EDUCATION", "MARRIAGE"]      # 원-핫 인코딩 대상
ordinal     = [f"PAY_{i}" for i in range(1,7)]      # 순서 의미 유지 → 통과
numeric     = [c for c in df.columns
               if (df[c].dtype != "O") and (c not in (categorical + ["target"] + ordinal))]

print("\n컬럼 그룹 요약")
print(" - categorical:", categorical)
print(" - ordinal    :", ordinal)
print(" - numeric    : (총", len(numeric), ") →", numeric[:8], ("..." if len(numeric)>8 else ""))

# 수치형 통계(요약) — 너무 길어지지 않게 일부만
if numeric:
    print("\nnumeric describe (상위 6개 컬럼만 표기)")
    display(df[numeric[:6]].describe().T)

# ---- 3) Stratified Split ----
X = df[categorical + ordinal + numeric]
y = df["target"].astype(int)

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=TEST_SIZE, stratify=y, random_state=RANDOM_STATE
)

print("\nSplit 결과")
print(" - X_train/X_test:", X_train.shape, X_test.shape)
print(" - y_train 비율:", (y_train.value_counts(normalize=True)*100).round(2).astype(str)+"%")
print(" - y_test  비율:", (y_test.value_counts(normalize=True)*100).round(2).astype(str)+"%")

# ---- 4) ColumnTransformer ----
def make_ohe():
    # OneHotEncoder 버전 호환 처리
    try:
        return OneHotEncoder(handle_unknown="ignore", sparse_output=False)
    except TypeError:
        return OneHotEncoder(handle_unknown="ignore", sparse=False)

ct = ColumnTransformer(
    transformers=[
        ("cat", make_ohe(), categorical),     # 명목형 → One-Hot
        ("num", StandardScaler(), numeric),   # 수치형 → 표준화
        ("ord", "passthrough", ordinal),      # 순서형 → 그대로
    ],
    remainder="drop"
)

Xt_train = ct.fit_transform(X_train)  # ← train으로만 학습
Xt_test  = ct.transform(X_test)

print("\n변환 후 shape (train/test):", Xt_train.shape, Xt_test.shape)

# === 변환 세부 정보 출력 ===
# 4-1) 원-핫 결과: 카테고리/더미 개수
ohe = ct.named_transformers_["cat"]
cat_feature_count = sum(len(c) for c in ohe.categories_)
print("\nOne-Hot 카테고리 목록")
for name, cats in zip(categorical, ohe.categories_):
    print(f" - {name}: {list(cats)} (총 {len(cats)})")
print(" - One-Hot 전체 더미 수:", cat_feature_count)

# 4-2) 스케일러 통계(평균/표준편차) — 일부만
if numeric:
    scaler = ct.named_transformers_["num"]
    stats = pd.DataFrame({
        "mean": scaler.mean_,
        "scale(std)": scaler.scale_
    }, index=numeric)
    print("\nStandardScaler 통계 (상위 6개만 표시)")
    display(stats.head(6))

# 4-3) 변환 후 피처명 추출
def get_feature_names(ct_: ColumnTransformer):
    names = []
    for name, trans, cols in ct_.transformers_:
        if name == "remainder" and trans == "drop":
            continue
        if hasattr(trans, "get_feature_names_out"):
            names.extend(list(trans.get_feature_names_out(cols)))
        elif trans == "passthrough":
            names.extend(list(cols if isinstance(cols, (list, tuple, np.ndarray)) else [cols]))
        else:
            names.extend(list(cols if isinstance(cols, (list, tuple, np.ndarray)) else [cols]))
    return np.array(names, dtype=object)

feat_names = get_feature_names(ct)
print("\n변환 피처 수:", len(feat_names))
print("피처명 미리보기:", feat_names[:20])

# 4-4) 변환 결과 샘플 행 확인
Xt_train_df = pd.DataFrame(Xt_train, columns=feat_names)
Xt_test_df  = pd.DataFrame(Xt_test,  columns=feat_names)
print("\n[변환된 Train 샘플 3행]")
display(Xt_train_df.head(3))

# ---- 4-5) 파일 저장 ----
Xt_train_df.to_csv("X_train_processed.csv", index=False)
Xt_test_df.to_csv("X_test_processed.csv", index=False)
y_train.to_csv("y_train.csv", index=False)
y_test.to_csv("y_test.csv", index=False)
print("\n저장 완료 → X_train_processed.csv, X_test_processed.csv, y_train.csv, y_test.csv")

# ---- 5) 리샘플링(옵션, train only) ----
if APPLY_SMOTE:
    sm = SMOTE(random_state=RANDOM_STATE)
    Xs_train, ys_train = sm.fit_resample(Xt_train, y_train)
    print("\nSMOTE 적용")
    print(" - 전 :", dict(y_train.value_counts()))
    print(" - 후 :", dict(pd.Series(ys_train).value_counts()))
    np.save("X_train_smote.npy", Xs_train)
    np.save("y_train_smote.npy", ys_train)
    print("저장: X_train_smote.npy, y_train_smote.npy")

if APPLY_CLUSTER_CENTROIDS:
    cc = ClusterCentroids(random_state=RANDOM_STATE)
    Xu_train, yu_train = cc.fit_resample(Xt_train, y_train)
    print("\nClusterCentroids 적용")
    print(" - 전 :", dict(y_train.value_counts()))
    print(" - 후 :", dict(pd.Series(yu_train).value_counts()))
    np.save("X_train_cc.npy", Xu_train)
    np.save("y_train_cc.npy", yu_train)
    print("저장: X_train_cc.npy, y_train_cc.npy")

# ---- 6) 빠른 시각 확인 ----
# (a) 클래스 불균형 — train/test
plt.figure()
vals_tr = y_train.value_counts().sort_index()
vals_te = y_test.value_counts().sort_index()
plt.bar(["train_0","train_1"], [vals_tr.get(0,0), vals_tr.get(1,0)])
plt.bar(["test_0","test_1"],  [vals_te.get(0,0), vals_te.get(1,0)])
plt.title("Class balance (train/test)")
plt.show()

# (b) 스케일 전후 히스토그램(임의 numeric 1개)
if len(numeric) > 0:
    feat = numeric[0]
    plt.figure(); plt.hist(X_train[feat], bins=50); plt.title(f"Original {feat} (train)"); plt.show()
    try:
        idx = np.where(feat_names == feat)[0][0]
        plt.figure(); plt.hist(Xt_train[:, idx], bins=50); plt.title(f"Standardized {feat} (train)"); plt.show()
    except Exception as e:
        print("표준화 컬럼 인덱스 탐색 실패:", e)
```

```python
# === 후처리: ID 드롭 + 피처명 저장 ===
import pandas as pd
import os

made = []

# 1) Train/Test에서 ID 제거한 버전 저장 (있을 때만)
for name in ["X_train_processed.csv", "X_test_processed.csv"]:
    if not os.path.exists(name):
        print(f" 파일 없음: {name}")
        continue
    df = pd.read_csv(name)
    had_id = "ID" in df.columns
    if had_id:
        df = df.drop(columns=["ID"])
        out = name.replace(".csv", "_noID.csv")
        df.to_csv(out, index=False)
        made.append(out)
        print(f" {name} -> {out} (ID 드롭, shape={df.shape})")
    else:
        print(f" {name} 에 ID 컬럼 없음 (shape={df.shape})")

# 2) 피처명 저장 (ID 드롭 버전 있으면 그 기준으로, 없으면 기존 train 기준)
feat_source = "X_train_processed_noID.csv" if os.path.exists("X_train_processed_noID.csv") else "X_train_processed.csv"
if os.path.exists(feat_source):
    cols = pd.read_csv(feat_source, nrows=0).columns
    cols.to_series().to_csv("feature_names.txt", index=False)
    made.append("feature_names.txt")
    print(f" feature_names.txt 저장 (기준: {feat_source}, {len(cols)}개 특성)")
else:
    print(" feature_names.txt 생성 실패: 기준 파일을 찾지 못함")

print("\n 생성/업데이트된 파일:", made if made else "(없음)")
```
